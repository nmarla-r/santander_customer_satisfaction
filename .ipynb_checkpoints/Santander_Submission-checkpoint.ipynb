

### Analysis and model building on Santander Customer Satisfaction dataset


Importing all the required modules 

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.model_selection import train_test_split # to split the dataset to train and validation data
from sklearn.feature_selection import VarianceThreshold # used to filter on threshold value
from sklearn.tree import DecisionTreeClassifier # sklearn Decision Tree module for model building
from sklearn.metrics import roc_curve, roc_auc_score # metrics for evaluating the model
from matplotlib import pyplot as plt
from sklearn.model_selection import GridSearchCV # for hyper parameter tuning
from sklearn.metrics import f1_score, make_scorer # metrics for evaluation
from imblearn.over_sampling import SMOTE # over samping module to deal with data imbalance
import xgboost as xgb # xgBoost module
from sklearn.metrics import confusion_matrix, accuracy_score # metrics for evaluation
```

Importing the kaggle train data from a csv file and assigning it to train_data and splitting it for validation dataset


```python
train_data = pd.read_csv('/content/train.csv', header=0)
train_data.head(10)
```

```python
# creating the x(variables) and y(target) from the data
X = train_data.drop('TARGET', axis=1)
Y = train_data['TARGET']
X.head()
```

```python
Y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

### EDA
Exploring the dataset to remove variables that do not provide any value to out model


```python
x_train.describe()
```

```python
# var3 seems to be having many rows with -999999 which is effecting the mean and std.dev 
# so these extreme values can be replaced with the median value to make it usable

x_train['var3'].replace(to_replace = -999999, value = 2, inplace = True)

# the row ID doesn't provide any value to our data so we can drop it

x_train.drop('ID', axis=1, inplace=True)
x_train.describe()
```

```python
# the number of variables provided are very high(370) but not all may contain relevent data so we have
# to do some feature selection to filter out the useful variables
# lets remove all columns with zero variance(constant)

vt = VarianceThreshold(threshold=0.0)
vt.fit(x_train)
constant_columns = [column for column in x_train.columns
                  if column not in x_train.columns[vt.get_support()]]
print(len(constant_columns))
```

constant_columns is a list of column names whose variance is 0 i.e. the value is constant for all rows hence do not provide any value to us 

```python
constant_columns
```

```python
x_train.info()
```

```python
# dropping all the columns that do not have any variance since they do not provide any value for the model
x_train.drop(constant_columns, axis=1, inplace=True)
x_train.info()
```

A total of 35 variable columns have been dropped

```python
# now we can also remove quasi constant variables since they also provide very little information as they re almost constant
# for this variance=0.05

vt_q = VarianceThreshold(threshold=0.05)
vt_q.fit(x_train)
quasi_columns = [column for column in x_train.columns
                  if column not in x_train.columns[vt_q.get_support()]]
print(len(quasi_columns))
```

Now, we drop quasi constant columns i.e. variables with very low variance and does not help in classifing out data. A total of 101 columns are dropped in this step

```python
x_train.drop(quasi_columns, axis=1, inplace=True)
x_train.info()
```

```python
# from 233 variable many can be highly correlated which doesn't provide any relevent information so we can keep one
# of the variables and delete the rest

cor_matrix = x_train.corr().abs()
print(cor_matrix)
```

```python
# selecting only the upper part of the correlation matrix as it is diagonally symetric

upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
print(upper_tri)
```

```python
# creating a list of columns to drop that are highly correlated from the upper_tri so that one column in a corellated relation stays and the other gets dropped

to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
len(to_drop)
```

A total of 71 columns are getting dropped here

```python
x_train.drop(to_drop, axis=1, inplace=True)
x_train.info()
```

performing all the same actions on validation dataset

```python
x_test['var3'].replace(to_replace = -999999, value = 2, inplace = True)
x_test.drop('ID', axis=1, inplace=True)
x_test.drop(constant_columns, axis=1, inplace=True)
x_test.drop(quasi_columns, axis=1, inplace=True)
x_test.drop(to_drop, axis=1, inplace=True)
x_test.info()
```

### Model 1

```python
clf1 = DecisionTreeClassifier(criterion = 'entropy')
```

```python
clf1.fit(x_train, y_train)
```

```python
clf1_pred = clf1.predict(x_test)
```

```python
#function to define evaluation metric for the predictions

def evaluate_model(dt_classifier):
    print("Train Accuracy :", accuracy_score(y_train, dt_classifier.predict(x_train)))
    print("Train Confusion Matrix:")
    print(confusion_matrix(y_train, dt_classifier.predict(x_train)))
    print("-"*50)
    print("Test Accuracy :", accuracy_score(y_test, dt_classifier.predict(x_test)))
    print("Test Confusion Matrix:")
    print(confusion_matrix(y_test, dt_classifier.predict(x_test)))
```

```python
evaluate_model(clf1)
```

```python
# looking at the roc_auc score of the model
clf1_prob = clf1.predict_proba(x_test)
roc_auc_score(y_test, clf1_prob[:,1])
```

```python
# function to plot roc
def plot_roc(predictions):
    fpr, tpr, _ = roc_curve(y_test, predictions[:,1])
    plt.clf()
    plt.plot(fpr, tpr)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC curve')
    plt.show()
```

```python
plot_roc(clf1_prob)
```

kaggle score for model 1 = 0.56390

### Model 2


```python
# only change is the criterion of the decision tree from entropy to gini

clf2 = DecisionTreeClassifier(criterion = 'gini', min_samples_split=2, splitter='best', random_state=42)

clf2.fit(x_train, y_train)

clf2_pred = clf2.predict(x_test)
```

```python
evaluate_model(clf2)
```

```python
clf2_prob = clf2.predict_proba(x_test)
roc_auc_score(y_test, clf2_prob[:,1])
```

```python
plot_roc(clf2_prob)
```

kaggle score for model 2 = 0.57472

### Imbalanced data problem
one major problem from this data set is that it is highly imbalanced hence the low score after kaggle submission of the above models         
To deal with this somewhat we can use oversampling

```python
# using SMOTE to balance the data

smote = SMOTE(random_state = 42)

X_res, y_res = smote.fit_resample(x_train, y_train)
print(y_train.value_counts())
print(y_res.value_counts())
```

### Model 3

```python
# training the new data on decision tree classifier

clf3 = DecisionTreeClassifier(criterion = 'gini', max_depth=10, random_state=42)
```

```python
clf3.fit(X_res, y_res)
```

```python
clf3_pred = clf3.predict(x_test)
```

```python
evaluate_model(clf3)
```

```python
clf3_prob = clf3.predict_proba(x_test)
roc_auc_score(y_test, clf3_prob[:,1])
```

```python
plot_roc(clf3_prob)
```

kaggle score for model 3 = 0.79127

```python
clf4_pred = clf4.predict(x_test)
```

### Hyper parameter optimization

```python
# Part of GridCV code that was run seperately for hyper parameter tuning of the XGBoost model

xgb_clf = xgb.XGBClassifier(n_estimators = 100, random_state = 42)
# hyper parameters

params = {'max_depth' : [3,5,7] , 
          'learning_rate' : [0.5, 0.1], 
          'gamma' :[0, 0.25,1.0],
          'reg_lambda' : [0, 1.0, 10.0],}

gridcv = GridSearchCV(xgb_clf, param_grid = params, cv = 3 )

gridcv.fit(X_res, y_res, early_stopping_rounds = 20, eval_metric = "auc", 
            eval_set = [(x_train, y_train), (x_test, y_test)] )

print('GridSearchCV best parameters : ', gridcv.best_params_)
```

### Model 5

```python
clf5 = xgb.XGBClassifier(max_depth=7, 
n_estimators=350, learning_rate=0.025, nthread=4, subsample=0.95,
colsample_bytree=0.85, seed=42 , gamma=1.0, reg_lambda=1.0)
```

```python
clf5.fit(X_res, y_res, early_stopping_rounds = 200, eval_metric = "auc", 
            eval_set = [(x_train, y_train), (x_test, y_test)] )
```

```python
clf5_prob = clf5.predict_proba(x_test)
roc_auc_score(y_test, clf5_prob[:,1])
```

```python
plot_roc(clf5_prob)
```

kaggle score for model 5 = 0.82742
